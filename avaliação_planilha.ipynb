```python
import requests
import json

API_URL = 'https://script.google.com/macros/s/AKfycbySnDjCqElRBHiQHI8K85pUeqPW7MsFeoPxvqVJVMbwxKtNBtjKcGAZVgCpxNn-K9oP/exec'  # Replace with your actual Google Apps Script web app URL

def enviar_mensagem(name, email):
    response = requests.post(
        API_URL,
        json={"name": name, "email": email}
    )
    print(response.text)

def enviar_varias_mensagens(messages):
    for message_dict in messages: # complete this line
        enviar_mensagem(name='name' , email='email')

def pegar_todas_mensagens():
    response = requests.get(API_URL) # complete this line
    data = response.json() # complete this line
    print(json.dumps(data, indent=2))

def pegar_ultima_mensagem():
    response = requests.get(API_URL) # complete this line
    data = response.json()
    if data:
        last = data[-1] # complete this line
        print(json.dumps(last, indent=2))
    else:
        print("Nenhuma mensagem achada.")

if __name__ == "__main__":
    print("Enviando uma mensagem...")
    enviar_mensagem("alice", "alice@example.com")

    print("\nEnviando varias mensagens...")
    messages = [
        {"name": "bob", "email": "bob@example.com"},
        {"name": "charlie", "email": "charlie@example.com"}
    ]
    enviar_varias_mensagens(messages)

    print("\nRecuperando todas as mensagens...")
    pegar_todas_mensagens()

    print("\nRecuperando ultima mensagem...")
    pegar_ultima_mensagem()
```
